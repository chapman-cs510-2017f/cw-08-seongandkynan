# Julia Benchmark
Using the numba library, we could run the same code 24 times faster. It is amazing to watch the difference, because we could run the same Python code without making many changes. Numba could run the code with high performace with just-in-time compiling the code to native machine language using LLVM (Low Level Virtual Machine) technology. For example, the first plot took 41 seconds, while the numba's vectorized version took only 1.65 seconds.

```python
import cplane_julia
%time jp = cplane_julia.JuliaPlane(0, -2, 2, 2000, -2, 2, 2000)
%time jp.show()
```

```python
%time jp = cplane_julia.JuliaPlane(-1, -2, 2, 2000, -2, 2, 2000)
%time jp.show()
```

```python
%time jp = cplane_julia.JuliaPlane(0.3, -2, 2, 2000, -2, 2, 2000)
%time jp.show()
```

```python
%time jp = cplane_julia.JuliaPlane(-1.037 + 0.17j, -2, 2, 2000, -2, 2, 2000)
%time jp.show()
```

```python
%time jp = cplane_julia.JuliaPlane(-0.624 + 0.435j, -2, 2, 2000, -2, 2, 2000)
%time jp.show()
```

```python
%time jp = cplane_julia.JuliaPlane(-0.8 + 0.2j, -2, 2, 2000, -2, 2, 2000)
%time jp.show()
```

```python
%time jp = cplane_julia.JuliaPlane(-0.835 - 0.2321j, -2, 2, 2000, -2, 2, 2000)
%time jp.show()
```
